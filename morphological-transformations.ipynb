````python
#  import library
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
# input 2 images for analysis 
image1 = cv.imread('data/R.jpeg', cv.IMREAD_GRAYSCALE)
image2 = cv.imread('data/OIP.jpeg', cv.IMREAD_GRAYSCALE)
````

# Erosion
<p style="color:#159364; font-family:cursive; font-size:20px;"><br>The basic idea of erosion is just like soil erosion only, it erodes away the boundaries of foreground object. A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel is 1, otherwise it is eroded. It is useful for removing small white noises, detach two connected objects etc.
</p>


````python
kernel = np.ones((9,9),np.uint8)
erosion = cv.erode(image1,kernel,iterations = 1)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

ax[0].imshow(cv.cvtColor(image1, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')

ax[1].imshow(cv.cvtColor(erosion, cv.COLOR_BGR2RGB))
ax[1].set_title('Erosion')

plt.subplots_adjust(wspace=0.5)

# show the subplots
plt.show()
````

# Dilation
<p style="color:#159364; font-family:cursive; font-size:20px;"><br>It is just opposite of erosion. Here, a pixel element is '1' if at least one pixel under the kernel is '1'.</p>

````python
dilation = cv.dilate(image1,kernel, iterations = 1) # use 9 x 9 kernel 

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

ax[0].imshow(cv.cvtColor(image1, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')

ax[1].imshow(cv.cvtColor(dilation, cv.COLOR_BGR2RGB))
ax[1].set_title('Dilation')

plt.subplots_adjust(wspace=0.5)

# show the subplots
plt.show()
````

# Opening
<p style="color:#159364; font-family:cursive; font-size:20px;"><br>Opening is just another name of erosion followed by dilation. It is useful in removing noise.</p>

````python
opening = cv.morphologyEx(image1, cv.MORPH_OPEN, kernel)

# create subplots to display the original image and its erosion side by side
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# display the original image in the left subplot
ax[0].imshow(cv.cvtColor(image1, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')

# display the erosion image in the right subplot
ax[1].imshow(cv.cvtColor(opening, cv.COLOR_BGR2RGB))
ax[1].set_title('Opening')

# adjust the spacing between the subplots
plt.subplots_adjust(wspace=0.3)

# show the subplots
plt.show()
````

# Closing
<p style="color:#159364; font-family:cursive; font-size:20px;"><br>Closing is reverse of Opening, Dilation followed by Erosion. It is useful in closing small holes inside the foreground objects, or small black points on the object.</p>

````python
closing = cv.morphologyEx(image1, cv.MORPH_CLOSE, kernel)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

ax[0].imshow(cv.cvtColor(image1, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')

ax[1].imshow(cv.cvtColor(closing, cv.COLOR_BGR2RGB))
ax[1].set_title('Closing')

plt.subplots_adjust(wspace=0.3)

# show the subplots
plt.show()
````

# Morphological Gradient
<p style="color:#159364; font-family:cursive; font-size:20px;"><br>It is the difference between dilation and erosion of an image.The morphological gradient is useful to detect the outline of an object. It can be used for edge detection.</p>

````python
gradient = cv.morphologyEx(image1, cv.MORPH_GRADIENT, kernel)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

ax[0].imshow(cv.cvtColor(image1, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')

ax[1].imshow(cv.cvtColor(gradient, cv.COLOR_BGR2RGB))
ax[1].set_title('Morphological Gradient')

plt.subplots_adjust(wspace=0.3)

# show the subplots
plt.show()
````
